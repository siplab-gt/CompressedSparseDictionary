```python
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import HTML
%matplotlib inline

from utils.dict_plotting import show_dict, show_phi_vid
```

### Load Data Files

```python
base_data_file = np.load('./results/traindata_05-30-2020_none.npz')
base_train_time = base_data_file['time']
base_dictionary_list = base_data_file['phi']
base_train_loss = base_data_file['train']
base_val_loss = base_data_file['val']

dbd_data_file = np.load('./results/traindata_05-15-2020_blockdiagonal.npz')
dbd_train_time = dbd_data_file['time']
dbd_dictionary_list = dbd_data_file['phi']
dbd_train_loss = dbd_data_file['train']
dbd_val_loss = dbd_data_file['val']

bd_data_file = np.load('./results/traindata_05-16-2020_band_diagonal.npz')
bd_train_time = bd_data_file['time']
bd_dictionary_list = bd_data_file['phi']
bd_train_loss = bd_data_file['train']
bd_val_loss = bd_data_file['val']
```

### Plot Training Loss per Epoch

```python
plt.figure(figsize=(8,8))
plt.plot(base_train_loss, label='Uncompressed')
plt.plot(dbd_train_loss, label='DBD J=4')
plt.plot(bd_train_loss, label='BD J=4')
plt.legend()
plt.title('Train Loss Values vs Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
```

### Plot Validation Loss per Epoch

```python
plt.figure(figsize=(8,8))
plt.semilogy(base_val_loss, label='Uncompressed')
plt.semilogy(dbd_val_loss, label='DBD J=4')
plt.semilogy(bd_val_loss, label='BD J=4')
plt.legend()
plt.title('Val Loss Values vs Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
```
